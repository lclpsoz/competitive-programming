```python
import matplotlib.pyplot as plt

def f (n, minus, adj):
    ret = 0
    for i in range (n):
        for j in range (i+minus):
            ret += adj[i+1]

maxE = 50000
maxV = 10000

labels = []
vals = []
for n in range (1, maxV):
    e = min (n-1, (2*maxE)//n)
    print (maxE//(), minus, f(maxE//e, e))
    labels.append (e)
    vals.append (f(maxE//e, e))
    print (e)
print ("here!")
print (labels)
print (vals)
plt.plot(range (len(vals)), vals)
plt.ylim(0, 7e8)
plt.xlim(223, 0)
plt.grid()
plt.show()
```
